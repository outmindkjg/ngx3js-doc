```python
from glob import glob
import re
import os
import json

def checkDir(parentPath : str, fileName : str) :
    os.makedirs('src/app/' + parentPath + '/' + fileName, exist_ok=True)


def loadFile(file_name) :
    with open(file_name, "r") as f :
        return "".join(f.readlines())

def saveFile(file_name, contents) :
    with open(file_name, "w") as f :
        f.write(contents)
defTs = loadFile('src/app/uw/test/test.component.ts')
defHtml = loadFile('src/app/uw/test/test.component.html')
defScss = loadFile('src/app/uw/test/test.component.scss')

def createCom(parentPath : str, fileName : str, pageName : str) :
    resultCom = defTs
    resultCom = resultCom.replace("test",fileName).replace("TestComponent", fileName[0].upper() + fileName[1:] + 'Component')
    resultCom = resultCom.replace("NAME",pageName)
    saveFile('src/app/' + parentPath + '/' + fileName + '/' + fileName + '.component.ts', resultCom)
    resultCom = resultCom.replace("../../shared/","../../../shared/")
    saveFile('src/app/design/' + parentPath + '/' + fileName + '/' + fileName + '.component.ts', resultCom)
    print('import {{ {1}Component }} from "./{0}/{0}.component";'.format(fileName, fileName[0].upper() + fileName[1:]))

def createHtml(parentPath : str, fileName : str, pageName : str) :
    resultHtml = defHtml
    resultHtml = resultHtml.replace("test",fileName)
    resultHtml = resultHtml.replace("NAME",pageName)
    saveFile('src/app/' + parentPath + '/' + fileName + '/' + fileName + '.component.html', resultHtml)
    saveFile('src/app/design/' + parentPath + '/' + fileName + '/' + fileName + '.component.html', resultHtml)

def createScss(parentPath : str, fileName : str, pageName : str) :
    resultScss = defScss
    resultScss = resultScss.replace("test",fileName)
    resultScss = resultScss.replace("NAME",pageName)
    saveFile('src/app/' + parentPath + '/' + fileName + '/' + fileName + '.component.scss', resultScss)
    saveFile('src/app/design/' + parentPath + '/' + fileName + '/' + fileName + '.component.scss', resultScss)

def getTemp(parentPath : str, fileName : str, pageName : str) :
    checkDir(parentPath, fileName)
    createCom(parentPath,fileName, pageName)
    createHtml(parentPath,fileName, pageName)
    createScss(parentPath,fileName, pageName)
```
